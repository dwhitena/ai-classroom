```
import spacy
import torch
from transformers import RobertaTokenizer, RobertaModel
```

# SpaCy tokenization and word embeddings

```
# Load the pretrained SpaCy English models.
nlp = spacy.load("en_core_web_lg")

# Tokenize and compute the embeddings for each token.
tokens = nlp("dog cat banana afskfsd")

# Print the embedding for dog.
print(tokens[0].vector)
# for token in tokens:
#     print(token.text, token.has_vector, token.vector_norm, token.is_oov)
```

# Hugging Face tokenization and word embeddings ðŸ¤—

```
# Import a pretrained tokenizer. This is from the Roberta model from Facebook AI.
# https://ai.facebook.com/blog/roberta-an-optimized-method-for-pretraining-self-supervised-nlp-systems/
tokenizer = RobertaTokenizer.from_pretrained('roberta-large')

# Import the pretrained Roberta model.
model = RobertaModel.from_pretrained('roberta-large')
```

```
# Get tokens.
tokens = tokenizer.encode("dog cat banana afskfsd", add_special_tokens=True)

input_ids = torch.tensor(tokens).unsqueeze(0)
outputs = model(input_ids)

outputs[0]
```
